```python
# load a CSV file using pandas and make “CustomerID” as an index

import pandas as pd

df_mall = pd.read_csv("Mall_Customers.csv", index_col = "CustomerID")
df_mall.head(3)
```

```python
# visualize the four columns of interest using seaborn's scatterplot

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(1, figsize = (10, 5))
sns.scatterplot(
    data = df_mall,
    x = 'Spending Score (1-100)',
    y = 'Annual Income (k$)',
    hue = 'Genre',
    size = 'Age',
    palette = 'Set2'
);
```

```python
# normalize the data to eliminate outlier or anomolies
# We are dropping the “Gender” and “Age” columns and will be using the rest of them to find the clusters. 
from sklearn import preprocessing

X = df_mall.drop(['Genre', 'Age'], axis = 1)
X_norm = preprocessing.normalize(X)
```

```python
# The optimal value of K in the K-Means algorithm can be found using the Elbow method. This involves finding the inertia value of every K number of clusters from 1-10 and visualizing it.

import numpy as np
from sklearn.cluster import KMeans

def elbow_plot(data, clusters):
    inertia = []
    for n in range(1, clusters):
        algorithm = KMeans(
            n_clusters = n,
            init = "k-means++",
            random_state = 125,
        ) 
        algorithm.fit(data)
        inertia.append(algorithm.inertia_)
# Plot the elbow 
    plt.plot(np.arange(1, clusters), inertia, 'o')
    plt.plot(np.arange(1, clusters), inertia, '-', alpha = 0.5)
    plt.xlabel('Number of Clusters'), plt.ylabel('Inertia') 
    plt.show() 

elbow_plot(X_norm, 10)
```

The resultsof the plot suggest that three is the optimal number of clusters.

```python
#  Use KMeans algorithm from scikit-learn and provide it the K value. After that we will fit it on our training dataset and get cluster labels.


algorithm = KMeans(n_clusters = 3, init='k-means++', random_state = 125)
algorithm.fit(X_norm)
labels = algorithm.labels_
```

```python
# visualize the clusters as scatterplots

sns.scatterplot(data = X, x = 'Spending Score (1-100)', y = 'Annual Income (k$)', hue = labels, palette = 'Set2')
```

```python
# using different features
X = df_mall.drop(['Genre', 'Annual Income (k$)'], axis = 1)
X_norm = preprocessing.normalize(X)
```

```python
elbow_plot(X_norm, 10)
```

```python
# train the algorithm with different features, same number of clusters
algorithm = KMeans(n_clusters = 3, init = 'k-means++' ,random_state = 125)
algorithm.fit(X_norm)
labels = algorithm.labels_
```

```python
# plot the clusters 
sns.scatterplot(data = X, x = 'Age', y = 'Spending Score (1-100)', hue = labels, palette = 'Set2')
```
